```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sb
from matplotlib import rcParams
import pandas as pd
import numpy as np
from math import *
```

```python
data = np.loadtxt('http://www.scipy-lectures.org/_downloads/populations.txt').transpose()
# информация в виде списка, data[0] - года
```

```python
# Задачи 1,2
```

```python
plt.figure(figsize=(15, 10))
plt.xlabel('Year', fontsize = 16)
plt.ylabel('Population', fontsize = 16, rotation = 90)

di = {1: 'hares', 2: 'lynxes', 3 : 'carrots'}
colors = dict(zip([i for i in range(1, 4)], ['blue', 'limegreen', 'red']))
colors_max = dict(zip([i for i in range(1, 4)], ['navy', 'darkgreen', 'crimson']))

for i in range(1, 4):
    plt.xticks = data[0]
    maximum_population = {data[0][list(data[i]).index(max(data[i]))] : max(data[i])}
    graph_data = dict(zip(data[0], data[i]))
    line = sb.lineplot(data = graph_data, label = di[i], color=colors[i])
    line.xticks = data[0]
    line.set(title='Population in each year')
    max_pops = sb.scatterplot(data = maximum_population, label = f'max {di[i]}', color = colors_max[i], sizes = 80, legend = 'brief')
plt.legend()
plt.show()
```

```python
# Задача 3
```

```python
plt.figure(figsize = (15, 10))
plt.xlabel('year', fontsize = 15)
plt.ylabel('percent of population', rotation = 90, fontsize = 15)

stack_dict = {data[0][i] : [floor((data[1][i]/(data[1][i] + data[2][i])) * 100)/100, (ceil(data[2][i]/(data[1][i] + data[2][i])*100))/100] for i in range(len(data[0]))}
for i in range(len(data[0])):
    plt.bar(data[0], data[1], color = 'blue', width = 0.8)
    plt.text(data[0][i]-0.38, data[1][i]/2 - 1000, f'{int(stack_dict[data[0][i]][0]*100)}%',color = 'white')
    plt.bar(data[0], data[2], bottom = data[1], color = 'limegreen', width = 0.8)
    plt.text(data[0][i]-0.38, data[1][i] + data[2][i]/2 - 1000, f'{int(stack_dict[data[0][i]][1]*100)}%', color = 'black')
plt.legend(['hares', 'lynxes'])
plt.show()
```

```python
# Задача 4
```

```python
mean_hare = np.mean(data[1])
std_hare = np.std(data[1])
mean_lynx = np.mean(data[2])
std_lynx = np.std(data[2])
mean_carrot = np.mean(data[3])
std_carrot = np.std(data[3])

fig, (fig1, fig2, fig3)  = plt.subplots(1, 3, figsize = (20, 8))

fig1.bar(data[0], data[1], label = 'hare', width = 0.8, color = 'blue')
fig1.plot(data[0], [mean_hare for i in range(len(data[0]))], label = 'mean', color = 'orange')
fig1.plot(data[0], [std_hare for i in range(len(data[0]))], label = 'std', color = 'lightblue')
fig1.set_title('hare')

fig2.bar(data[0], data[2], label = 'lynx', width = 0.8, color = 'limegreen')
fig2.plot(data[0], [mean_lynx for i in range(len(data[0]))], label = 'mean', color = 'orange')
fig2.plot(data[0], [std_lynx for i in range(len(data[0]))], label = 'std', color = 'lightblue')
fig2.set_title('lynx')

fig3.bar(data[0], data[3], label = 'carrot', width = 0.8, color = 'red')
fig3.plot(data[0], [mean_carrot for i in range(len(data[0]))], label = 'mean', color = 'orange')
fig3.plot(data[0], [std_carrot for i in range(len(data[0]))], label = 'std', color = 'lightblue')
fig3.set_title('carrot')

fig1.legend()
fig1.plot()
fig2.legend()
fig2.plot()
fig3.legend()
fig3.plot()
```
